```python
import pandas as pd
import polars as pl
import numpy as np
import os
from tqdm.auto import tqdm
from matplotlib import pyplot as plt
import pickle

from sklearn.metrics import r2_score
from lightgbm import LGBMRegressor
import lightgbm as lgb
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.ensemble import VotingRegressor

import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_columns = None
```

```python
class CONFIG:
    seed = 42
    target_col = "responder_6"
    feature_cols = ["symbol_id", "time_id"] \
        + [f"feature_{idx:02d}" for idx in range(79)] \
        + [f"responder_{idx}_lag_1" for idx in range(9)]
    categorical_cols = []
```

```python
# To reduce memory usage, we use Polars to load the data and downcast float64 to float32.
print("Loading and preprocessing data with Polars for memory efficiency...")

# Scan the parquet files (lazy loading)
train_lazy = pl.scan_parquet("../jane-street-real-time-market-data-forecasting_Dataset/training.parquet")
valid_lazy = pl.scan_parquet("../jane-street-real-time-market-data-forecasting_Dataset/validation.parquet")

# Get schema to identify float64 columns for downcasting
schema = train_lazy.schema
dtype_map = {col: pl.Float32 for col, dtype in schema.items() if dtype == pl.Float64}

# Apply casting to downcast float64 to float32
train_lazy = train_lazy.cast(dtype_map)
valid_lazy = valid_lazy.cast(dtype_map)

# Load training data into pandas
print("Loading training data...")
train = train_lazy.collect().to_pandas()
print(f"Training data shape: {train.shape}")

# Load validation data into pandas
print("Loading validation data...")
valid = valid_lazy.collect().to_pandas()
print(f"Validation data shape: {valid.shape}")

train.shape, valid.shape
```

```python
# The step of merging the validation set into the training set has been removed
# to follow a standard validation procedure.
# This cell is now intentionally left blank.
# Original code:
# # Trick of boosting LB score: 0.45->0.49
# train = pd.concat([train, valid]).reset_index(drop=True)
# train.shape
```

```python
def get_model(seed):
    # XGBoost parameters for CPU training
    # Reverting to CPU training for stability after encountering persistent GPU errors.
    XGB_Params = {
        'learning_rate': 0.05,
        'max_depth': 6,
        'n_estimators': 200,
        'subsample': 0.8,
        'colsample_bytree': 0.8,
        'reg_alpha': 1,
        'reg_lambda': 5,
        'random_state': seed,
        # 'device' parameter is removed to default to stable CPU training.
    }
    
    XGB_Model = XGBRegressor(**XGB_Params)
    return XGB_Model
```

```python
X_train = train[ CONFIG.feature_cols ]
y_train = train[ CONFIG.target_col ]
w_train = train[ "weight" ]
X_valid = valid[ CONFIG.feature_cols ]
y_valid = valid[ CONFIG.target_col ]
w_valid = valid[ "weight" ]

# --- GPU Compatibility Check & Fix ---
# Forcing all feature columns to float32 and integer columns to int32 
# can improve compatibility with the Metal backend.
print("Converting data types for GPU compatibility...")
for col in X_train.columns:
    if X_train[col].dtype == 'float64':
        X_train[col] = X_train[col].astype('float32')
        X_valid[col] = X_valid[col].astype('float32')
    if X_train[col].dtype == 'int64':
        X_train[col] = X_train[col].astype('int32')
        X_valid[col] = X_valid[col].astype('int32')

print("Data type conversion complete.")
# ------------------------------------

X_train.shape, y_train.shape, w_train.shape, X_valid.shape, y_valid.shape, w_valid.shape
```

```python
%%time
model = get_model(CONFIG.seed)
# We add an evaluation set and the verbose parameter to monitor the training progress.
# The model will print the evaluation metric (default is RMSE for regression) on the validation set every 10 rounds.
model.fit(
    X_train, y_train, 
    sample_weight=w_train,
    eval_set=[(X_valid, y_valid)],
    verbose=1
)
```

```python
y_pred_train1 = model.predict(X_train.iloc[:X_train.shape[0]//2])
y_pred_train2 = model.predict(X_train.iloc[X_train.shape[0]//2:])
train_score = r2_score(y_train, np.concatenate([y_pred_train1, y_pred_train2], axis=0), sample_weight=w_train )
train_score
```

```python
y_pred_valid = model.predict(X_valid)
valid_score = r2_score(y_valid, y_pred_valid, sample_weight=w_valid )
valid_score
```

```python
y_means = { symbol_id : -1 for symbol_id in range(39) }
for symbol_id, gdf in train[["symbol_id", CONFIG.target_col]].groupby("symbol_id"):
    y_mean = gdf[ CONFIG.target_col ].mean()
    y_means[symbol_id] = y_mean
    print(f"symbol_id = {symbol_id}, y_means = {y_mean:.5f}")
```

```python
cv_detail = { symbol_id : 0 for symbol_id in range(39) }
for symbol_id, gdf in valid.groupby("symbol_id"):
    X_valid = gdf[ CONFIG.feature_cols ]
    y_valid = gdf[ CONFIG.target_col ]
    w_valid = gdf[ "weight" ]
    y_pred_valid = model.predict(X_valid)
    score = r2_score(y_valid, y_pred_valid, sample_weight=w_valid )
    cv_detail[symbol_id] = score
    
    print(f"symbol_id = {symbol_id}, score = {score:.5f}")
```

```python
sids = list(cv_detail.keys())
plt.bar(sids, [cv_detail[sid] for sid in sids])
plt.grid()
plt.xlabel("symbol_id")
plt.ylabel("CV score")
plt.show()
```

**Save Result**

```python
result = {
    "model" : model,
    "cv" : valid_score,
    "cv_detail" : cv_detail,
    "y_mean" : y_means,
}
with open("result.pkl", "wb") as fp:
    pickle.dump(result, fp)
```
