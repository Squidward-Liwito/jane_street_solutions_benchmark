```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'SimHei', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# 数据
data = {'天气': ['晴','晴','阴','雨','雨','雨','阴','晴','晴','雨','晴','阴','阴','雨'],
        '温度': ['高','高','高','高','低','低','低','高','低','低','低','高','低','高'],
        '风力': ['小','大','小','小','小','大','大','小','小','小','大','大','小','大'],
        '玩球': ['否','否','是','是','是','否','是','否','是','是','是','是','是','否']}

df = pd.DataFrame(data)

# 特征编码
X = pd.get_dummies(df[['天气','温度','风力']])
y = df['玩球'].map({'否':0,'是':1})

# 创建决策树分类器
clf = DecisionTreeClassifier(criterion='entropy')  # 或 'gini'
clf = clf.fit(X, y)

# 可视化
plt.figure(figsize=(25,15))  # 宽20，高10
tree.plot_tree(clf, feature_names=X.columns, class_names=['否','是'], filled=True)
```
